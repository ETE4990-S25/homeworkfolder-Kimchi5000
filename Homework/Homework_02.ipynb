````python
#Parent class
class Animals:
    def __init__(self, name, age):
        self.name = name
        self.age = age

#Subclasses for pets and farm animals
class Pet(Animals):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.friend = True
        self.eat = False

class Farm(Animals):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.eat = True
        self.helper = True


#Subclasses for specific type of animal
class Cat(Pet):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.friend = True
        self.eat = False
        self.type = "Maine Coone"

class Dog(Pet):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.friend = True
        self.eat = False
        self.type = "Corgie"

class Chicken(Farm):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.eat = True

class Pig(Farm):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.eat = True
  
class Cow(Farm):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.eat = True

class Horse(Farm):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.eat = False

listOfAnimals = [Cat("Mitsy", 3), Dog("Tex", 3), Horse("Grumpy", 5)]


#Functions to manage farm
class FarmManagement:
    def __init__(self):
        self.animals = listOfAnimals

    def display_animals(self):
        for animal in self.animals:
            print(f"Name:{animal.name.title()}, Age:{animal.age}, Eat:{animal.eat}")

    def add_animal(self, animal_type, name, age):
        animal_classes = {"cat": Cat,"dog": Dog, "chicken": Chicken, "pig": Pig, "cow": Cow, "horse": Horse}
        if animal_type in animal_classes:
           new_animal = animal_classes[animal_type](name, age)
           self.animals.append(new_animal)
           print(f"{name} has been added to the farm.")

    def remove_animal(self, name):
        for animal in self.animals:
            if animal.name.lower() == name.lower():
                self.animals.remove(animal)
                print(f"{name} has been removed from the farm.")
                return
            
            
#User menu for farm management
def user_input():
    farm = FarmManagement()

    while True:
        print("\nMain Menu:")
        print("1. Add an animal")
        print("2. Remove an animal")
        print("3. Display all animals")
        print("4. Exit")
        choice = input("Enter a Number(1-4):")

        if choice == "1":
            animal_type = input("Enter the type of animal (cat, dog, chicken, pig, cow, or horse):")
            name = input("Enter the animal's name:")
            age = int(input("Enter the animal's age:"))
            farm.add_animal(animal_type, name, age)

        elif choice == "2":
            name = input("Enter the animal's name that you would like to remove:")
            farm.remove_animal(name)

        elif choice == "3":
            farm.display_animals()

        elif choice == "4":
            print("Exiting Program")
            break

user_input()
            
        





````
