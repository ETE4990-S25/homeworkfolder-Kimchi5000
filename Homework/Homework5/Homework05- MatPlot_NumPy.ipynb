````python
import numpy as np

#Using genfromtxt() instead of loadtxt() because of formatting errors in csv.  Kept getting errors on line 179 column 4. 
vglist = np.genfromtxt(
    "vgsales.csv",
    dtype={'names': ('name', 'year', 'NA_Sales'), 'formats': ('U64', 'i', 'f')}, #Specifies the names and formats of the columns.
    delimiter=',', #Specifies that the csv file is separated by commas
    usecols=(1, 3, 6),
    max_rows=500, #Setting a max row limit because of errors; "invalid raise".  Some rows seem to have fewer columns
)

vglist.sort(order=['name', 'year', 'NA_Sales'])


print(vglist)
````
